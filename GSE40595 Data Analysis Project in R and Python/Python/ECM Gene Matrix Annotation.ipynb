```python
# import important modules
import numpy as np
import pandas as pd
# import matrix gene csv
matrix = pd.read_csv("../Data/Comparisons/Matrisome_Hs_MasterList.csv") 
matrix.head(20)

```

```python
MOCSvsMOSE = pd.read_csv("../Data/Comparisons/Sample_Comparisons/G1 - G2 MOCS vs MOSE.csv") # MOCSvsMOSE
MOCSvsMOSE_annot =pd.merge(MOCSvsMOSE,matrix,on="Gene.symbol", how="left") # merge dataframes based on gene name join
MOCSvsMOSE_annot = MOCSvsMOSE_annot.replace(np.nan,'NA', regex=True) # blank cell are renamed "NA"
MOCSvsMOSE_annot.to_csv(r'../Python/ECM annotation/MOCSvsMOSE_annot.csv', index = False) # export as csv
MOCSvsMOSE_annot.head(20)
```

```python
MOCSvsMOTEC = pd.read_csv("../Data/Comparisons/Sample_Comparisons/G1 - G3 MOCS vs MOTEC.csv") # MOCSvsMOTEC
MOCSvsMOTEC_annot =pd.merge(MOCSvsMOTEC,matrix,on="Gene.symbol", how="left") # merge dataframes based on gene name join
MOCSvsMOTEC_annot = MOCSvsMOTEC_annot.replace(np.nan,'NA', regex=True) # blank cell are renamed "NA"
MOCSvsMOTEC_annot.to_csv(r'../Python/ECM annotation/MOCSvsMOTEC_annot.csv', index = False) # export as csv
MOCSvsMOTEC_annot.head(20)
```
